```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return -x**2 * np.exp(-(x**2 + y**2))

x = np.linspace(-2.8, 2.8, 70)
y = np.linspace(-2.8, 2.8, 70)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

dx, dy = np.gradient(Z, x, y)
u, v = -dx, -dy

fig = plt.figure(figsize=(15, 10))
sur3 = fig.add_subplot(121, projection='3d')
sur3.plot_surface(X, Y, Z, cmap='plasma', edgecolor='black', alpha=0.6)
sur3.contour(X, Y, Z, zdir='z', offset=-0.4, cmap='coolwarm')
sur3.set_xlabel('X')
sur3.set_ylabel('Y')
sur3.set_zlabel('f(X,Y)')
sur3.set_title("3D ")

plot2d = fig.add_subplot(122)
contour = plot2d.contour(X, Y, Z, cmap='coolwarm')
plot2d.quiver(X, Y, u, v, color='green')
plot2d.set_xlabel('X')
plot2d.set_ylabel('Y')
plot2d.set_title("contour")

plt.tight_layout()
plt.show()
```
